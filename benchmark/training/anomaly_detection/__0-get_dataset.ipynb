```python
import os
import zipfile
import urllib.request
from tqdm import tqdm

# 設定 URL 和檔案名稱
URL1 = "https://zenodo.org/record/3678171/files/dev_data_ToyCar.zip?download=1"
ZIPFILE1 = "dev_data_ToyCar.zip"

URL2 = "https://zenodo.org/record/3727685/files/eval_data_train_ToyCar.zip?download=1"
ZIPFILE2 = "eval_data_train_ToyCar.zip"

# 創建資料夾
os.makedirs("dev_data", exist_ok=True)

# 自訂下載回呼函數以顯示進度條和檔案大小
def download_and_extract(url, zipfile_name, target_dir):
    # 取得檔案大小
    with urllib.request.urlopen(url) as u:
        file_size = int(u.getheader('Content-Length'))

    # 下載檔案並顯示進度條
    with tqdm(total=file_size, unit='B', unit_scale=True, miniters=1, desc=zipfile_name) as t:
        def reporthook(block_num, block_size, total_size):
            t.update(block_size)
        urllib.request.urlretrieve(url, zipfile_name, reporthook=reporthook)

    # 解壓檔案
    with zipfile.ZipFile(zipfile_name, 'r') as zip_ref:
        zip_ref.extractall(target_dir)

    # 刪除下載的 zip 檔案
    # os.remove(zipfile_name)
# 下載並解壓 dev_data_ToyCar.zip
download_and_extract(URL1, ZIPFILE1, "dev_data")
# 下載並解壓 eval_data_train_ToyCar.zip
download_and_extract(URL2, ZIPFILE2, "dev_data")
```
