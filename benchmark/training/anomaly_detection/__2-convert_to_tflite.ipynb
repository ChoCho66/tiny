```python
import pandas as pd
import os

result_folder = "./trained_models/ToyCar/baseline_tf23/result"

result_files = [
    'result.csv',
    'tflite_result.csv',
    'tflite_quant_result.csv',
    'tflite_quant_fullint_result.csv'
]

# 讀取資料檔案
file_paths = [os.path.join(result_folder,f) for f in result_files]

# 用來儲存每個檔案的 AUC 和 pAUC 結果
data = []

# 逐一處理每個檔案
for file in file_paths:
    with open(file, 'r') as f:
        lines = f.readlines()
        # 只取得最後一行 "Average" 的資料
        avg_line = lines[-2].strip().split(',')
        auc = float(avg_line[1])
        p_auc = float(avg_line[2])
        data.append({
            'File': file.split('/')[-1],  # 只取檔案名稱
            'AUC': auc,
            'pAUC': p_auc
        })

# 將資料轉換為 DataFrame
df = pd.DataFrame(data)

print(result_folder)
# 顯示比較結果
display(df)

# 以 AUC 由大到小排列
print()
print()
print()
print("以 AUC 由大到小排列")
df_sorted_AUC = df.sort_values(by='AUC', ascending=False)

# 顯示排序後的結果
display(df_sorted_AUC)

# 以 pAUC 由大到小排列
print()
print()
print()
print("以 pAUC 由大到小排列")
df_sorted_AUC = df.sort_values(by='pAUC', ascending=False)

# 顯示排序後的結果
display(df_sorted_AUC)
```
